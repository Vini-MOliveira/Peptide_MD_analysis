```python
N_res = 39
dfs_cont = []
N_rep = 3

for i in range(N_rep):
    for j in range(N_res):
        if j < 10:
            aux = g.glob('{}/desmond_md_job_trimer_phenol_60_{}/CONTACTS_C/res_cont_0{}.csv'.format(path,i+1,j))
        else:
            aux = g.glob('{}/desmond_md_job_trimer_phenol_60_{}/CONTACTS_C/res_cont_{}.csv'.format(path,i+1,j))
        df = pd.read_csv(aux[0])
        dfs_cont.append([i,j+1,df])
```

```python
contacts_per_residue = []
for df in (dfs_cont):
    for i in range(N_rep):
    # EACH Replica
        if df[0] == i:
            for j in range(N_res):
            #EACH residue
                resid = j+1
                if df[1] == resid:
                    s=0
                # EACH FRAME IN THE SPECIFIC REPLICA
                    for idx, row in df[2].iterrows():
                        if df[2]['Frame_No.'].iloc[idx] > 250:
                            s += (df[2]['Residue_Total'].iloc[idx])
                    print(resid,s/N_frames)
                    contacts_per_residue.append([resid,s/N_frames])
                
```

```python
res = []
cont_avg = []
cont_error = []
for j in range(N_res):
    resid = j+1
    su = 0
    sum_cont = []
    for i in contacts_per_residue:
        #print(i[0],i[1])
        if i[0] == resid:
            su += i[1]
            sum_cont.append(i[1])
    print(resid,su/N_rep,statistics.pstdev(sum_cont))
    res.append(resid)
    cont_avg.append(su/N_rep)
    cont_error.append(sem(sum_cont))
```

```python
fig = plt.figure(figsize=(4,3))

# creating the data values for the vertical y and horisontal x axis
x = np.array(res)
y = np.array(cont_avg)
error = np.array(cont_error)
# Build the plot
fig, ax = plt.subplots()
ax.bar(x, y, yerr=error, align='center', alpha=0.5, ecolor='black', capsize=3)
ax.yaxis.grid(True)

#ax.legend(loc="best")
ax.set_xlabel("Residues")
ax.set_ylabel('Contact freq (< 4.5 A)')

# to show our graph
plt.show()
```
